```python
#두번째줄까지 실행하고나면 이제 EDA시작 ㄱㄱ
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#여기까진 친절히 써드릴게 ^^
data = pd.read_csv('/gdrive/My Drive/DS_contest/trainSet.csv')
```

```python
data2=pd.read_csv('/gdrive/My Drive/DS_contest/data2.csv')
```

```python
datasix=pd.read_csv('/gdrive/My Drive/DS_contest/datasix.csv')
```

```python
data2.rename(columns={'Unnamed: 0':'index'},inplace=True)
data2.rename(columns={'Unnamed: 0.1':'ID index'},inplace=True)
data2
data_ = []
for i in range(232):
  data_.append(data2[data2['ID index']==i])
```

```python
data_x = []
for i in range(232):
  data_x.append(data2[(data2['ID index']==i) & (data2['X_t']<=500) & (data2['J_t']>400)])
```

```python
data2['normal'] = (data2['X_t']<=500) & (data2['J_t']>400)
data2 = data2.astype({'normal':'int'})
```

```python
plt.figure(figsize = (80,160))
for i in range(232):
  ax1 = plt.subplot(29,8,i+1)
  plt.title('%d index' %(i+1),fontsize=15)
  ax1.set_ylabel('X_t')
  ax1.plot(data_[i]['time'],data_[i]['X_t'], color='green')
  ax2=ax1.twinx()
  ax2.set_ylabel('Y_t')
  ax2.plot(data_[i]['time'],data_[i]['Y_t'], color='deeppink')

plt.tight_layout()
#plt.savefig('/gdrive/My Drive/DS_contest/Figures/Y_t,X_t-time.png')
plt.show
```
