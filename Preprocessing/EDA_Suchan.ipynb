```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as plot
import datetime as dt
import math
```

```python
df1 = pd.read_csv('/gdrive/My Drive/DS_contest/Finaldata1.csv')
df1.columns
```

```python
data_SFo = []
data_SFw = []
data_SFs = []
num_SFo = []
num_SFw = []
num_SFs = []

data_PCc = []
data_PCs = []
num_PCc = []
num_PCs = []

for i in range(len(df1)):
  if df1['SF_oil'][i] == 1:
    data_SFo.append(df1['Y_first6'][i])
    num_SFo.append(i)

for i in range(len(df1)):
  if df1['SF_water'][i] == 1:
    data_SFw.append(df1['Y_first6'][i])
    num_SFw.append(i)

for i in range(len(df1)):
  if df1['SF_slickwater'][i] == 1:
    data_SFs.append(df1['Y_first6'][i])
    num_SFs.append(i)

for i in range(len(df1)):
  if df1['PC_ceramic'][i] == 1:
    data_PCc.append(df1['Y_first6'][i])
    num_PCc.append(i)

for i in range(len(df1)):  
  if df1['PC_sand'][i] == 1:
    data_PCs.append(df1['Y_first6'][i])
    num_PCs.append(i)
```

```python
data1 = [data_SFo, data_SFw, data_SFs]
plt.boxplot(data1)
plt.title("stimulation fluid box plot")
plt.show()
```

```python
data2 = [data_PCc, data_PCs]
plt.boxplot(data2)
plt.title("proppant composition box plot")
plt.show()
```

```python
plt.plot(num_SFo, data_SFo, 'o', c = 'r', alpha = 0.3)
plt.plot(num_SFs, data_SFs, 'o', c = 'g', alpha = 0.3)
plt.plot(num_SFw, data_SFw, 'o', c = 'b', alpha = 0.3)
plt.show()
```

```python
plt.plot(num_PCc, data_PCc, 'o', c = 'c', alpha = 0.5)
plt.plot(num_PCs, data_PCs, 'o', c = 'y', alpha = 0.5)
plt.show()
```
