```python
#data1[i]에 각 행의 정보를 넣음
data1 = []
for i in range(280):
  data1.append(data.iloc[i])
  data1[i].dropna(inplace=True)
```

```python
#df1은 기본데이터와 처음6개월간 생산량 데이터만 포함하고있음
df1 = data1[0].to_frame().transpose()
for i in range(279):
  df1 = pd.concat([df1,data1[i+1].to_frame().transpose()],axis=0,sort=False, join='inner')
df1
```

```python
df1 = df1.drop(['CPA Pretty Well ID','On Prod YYYY/MM/DD','First Prod YYYY/MM','Last Prod. YYYY/MM'],axis=1)
```

```python
#이런식으로 저장하면됨
# df1.to_csv('/gdrive/My Drive/DS_contest/df1.csv')
```

```python
df1.rename(columns={'First 6 mo. Avg. GAS (Mcf)':'Y_first6'},inplace=True)
```

```python
df1
```

```python
#length 232(i=0 to 231)의 data2 list 만들기 (각각의 원소는 series)
data2 = []
for i in range(48,280):
  data2.append(data.iloc[i])
data2[0]
```

```python
#GAS_MONTH_1 ~ 36, Last 6 mo. Avg. GAS (Mcf), CND_MON_1 ~ 36, HRS_MONTH_1 ~ 36
#GAS , CND, HRS, time (1~36)]

#arb[i]는 series, data2[i], arbdf[i]는 데이터프레임
arb = []
arbdf = []
for i in range(232):
  arb.append(data2[i])
  data2[i] = data2[i].to_frame().transpose()
  arbdf.append(data2[i])
data2[0]
```

```python
for i in range(232):
  for j in range(35):
    data2[i] = pd.concat([data2[i],arbdf[i]],axis=0,sort=False, join='outer')
data2[0]
```

```python
# 32~67 GAS, 68:Last 6 mo. Avg. GAS (Mcf), 69~104 CND, 105~140 HRS
for i in range(232):
  GAS_list = []
  CND_list = []
  HRS_list = []
  for j in range(36):
    GAS_list.append(arb[i][j+32])
    CND_list.append(arb[i][j+69])
    HRS_list.append(arb[i][j+105])
  data2[i]['Y_t'] = GAS_list
  data2[i]['C_t'] = CND_list
  data2[i]['J_t'] = HRS_list
data2[0]
```

```python
for i in range(232):
  for j in range(36):
    data2[i] = data2[i].drop('GAS_MONTH_'+str(j+1),axis=1)
    data2[i] = data2[i].drop('CND_MONTH_'+str(j+1),axis=1)
    data2[i] = data2[i].drop('HRS_MONTH_'+str(j+1),axis=1)
data2[0]
```

```python
import datetime as dt
```

```python
for i in range(232):
  time_list = []
  # time_list의 첫항은 Last month - on month + 1
  time_list.append((dt.datetime.strptime(arb[i]['Last Prod. YYYY/MM'],'%Y/%m/%d') - dt.datetime.strptime(arb[i]['On Prod YYYY/MM/DD'],'%Y/%m/%d') - dt.timedelta(days=1095)).days // 30 + 1)
  for j in range(35):
    time_list.append(time_list[j]+1)
  data2[i]['time'] = time_list
data2[0]
```

```python
arbdf[0]
```

```python
for i in range(232):
  data2[i] = data2[i].drop('On Prod YYYY/MM/DD',axis=1)
  data2[i] = data2[i].drop('First Prod YYYY/MM',axis=1)
  data2[i] = data2[i].drop('Last Prod. YYYY/MM',axis=1)
data2[0]
```

```python
#df2는 2번생산정만 포함, 각 생산정마다 36row 가짐.
df2 = data2[0]
for i in range(231):
  df2 = pd.concat([df2,data2[i+1]],axis=0,sort=False, join='outer')
df2
```

```python
df2.rename(columns={'First 6 mo. Avg. GAS (Mcf)':'Y_first6'},inplace=True)
df2.rename(columns={'Last 6 mo. Avg. GAS (Mcf)':'Y_last6'},inplace=True)
df2
```

```python
df2.to_csv('/gdrive/My Drive/DS_contest/df2.csv')
```
